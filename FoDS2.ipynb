```python
import pandas as pd
import numpy as np

# 1. Create a student dataset
data = {
    'Student_ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [20, 21, 19, 22, 20],
    'Gender': ['F', 'M', 'M', 'M', 'F'],
    'Study_Hours': [15, 10, 12, 5, 8],
    'Failures': [0, 1, 0, 2, 0],
    'Final_Grade': [88, 72, 85, 55, 90]
}

# Create DataFrame
df = pd.DataFrame(data)

# 2. Identify the features in the dataset
print("Features in the dataset:")
features = df.columns
print(features)

# 3. Identify the datatype of each feature
print("\nData types of each feature:")
data_types = df.dtypes
print(data_types)

# 4. Check the five-point summary of the features
print("\nFive-point summary of numerical features:")
numerical_summary = df.describe(include=[np.number])
print(numerical_summary)

print("\nFive-point summary of categorical features:")
categorical_summary = df.describe(include=[object])
print(categorical_summary)

# 5. Explanation
print("\nExplanation:")
print("The dataset consists of student records with various attributes.")
print("Numerical features include 'Age', 'Study_Hours', 'Failures', and 'Final_Grade'. Categorical features include 'Student_ID', 'Name', and 'Gender'.")
print("The five-point summary for numerical features provides insights into the distribution and spread of these attributes, such as the minimum, maximum, and median values.")
print("The categorical summary shows the frequency of each category within categorical features.")
print("Actions to consider as a data scientist:")
print("- Data Cleaning: Ensure there are no missing or inconsistent values.")
print("- Feature Engineering: Create or modify features if necessary to enhance the dataset.")
print("- Exploratory Data Analysis (EDA): Visualize and analyze the relationships between features to gain insights into factors affecting student performance.")
print("- Modeling: Use statistical or machine learning models to predict outcomes, such as predicting the final grade based on study hours and other features.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame(data)
```

```python
print(df['Gender'].value_counts())
```

```python
sorted_df = df.sort_values(by='Final_Grade', ascending=False)
print(sorted_df)
```

```python
print(df['Name'].nunique())
```

```python
df['Performance'] = np.where(df['Final_Grade'] >= 85, 'High', 'Low')
print(df.head())
```

```python
plt.hist(df['Final_Grade'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.scatter(df['Study_Hours'], df['Final_Grade'], color='green')
plt.title('Study Hours vs. Final Grade')
plt.xlabel('Study Hours')
plt.ylabel('Final Grade')
plt.show()
```

```python
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.barplot(x='Gender', y='Final_Grade', data=df, palette='viridis')
plt.title('Average Final Grade by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade')
plt.show()
```

Final Grades: The histogram reveals the distribution of grades, showing most students score between 70 and 90.

Study Hours and Performance: The scatter plot suggests a positive relationship between study hours and final grades, though it's not perfectly linear.

Gender Impact:The bar plots show that the dataset has more males than females, but females tend to score slightly higher on average.
